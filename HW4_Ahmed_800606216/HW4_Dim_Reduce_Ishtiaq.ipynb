# Import 

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
import time
from sklearn.metrics import precision_score, recall_score, f1_score
from sklearn.metrics import accuracy_score
from sklearn.datasets import fetch_openml
from sklearn.utils import check_random_state
```

## IRIS Dataset : Import and Transform

```python
iris = datasets.load_iris()
X = iris.data[:, 0:4]
y = iris.target

# train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1, stratify=y)

# Standardization

sc = StandardScaler()
X_train_std = sc.fit_transform(X_train)
X_test_std = sc.transform(X_test)
```

## IRIS Dataset: PCA

```python
print('Iris Dataset: PCA\n')

# Perform PCA

from sklearn.decomposition import PCA

pca = PCA() 
X_train_pca = pca.fit_transform(X_train_std)
X_test_pca = pca.transform(X_test_std)

# Plotting Explained variance Ratio
plt.figure(figsize=(8, 6))
plt.bar(range(1, 5), pca.explained_variance_ratio_, alpha=0.5, align='center', label = 'Individual Explained variance')
plt.step(range(1, 5), np.cumsum(pca.explained_variance_ratio_), where='mid', label = 'Cumulative Explained variance')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.yticks(np.arange(0,1,0.1))
plt.title('Explained Variance Ratio Plot of PCA')
plt.legend()
plt.show()

# Reduce the data to 2 dimensions

pca = PCA(n_components=2) 
X_train_pca = pca.fit_transform(X_train_std)
X_test_pca = pca.transform(X_test_std)


# Creating multiple instance of decission tree using different hyper paramter

dtree1 = DecisionTreeClassifier(criterion='gini', max_depth=4, random_state=1)
dtree2 = DecisionTreeClassifier(criterion='gini', max_depth=40, random_state=1)
dtree3 = DecisionTreeClassifier(criterion='gini', max_depth=400, random_state=1)

# Decission tree classifier 1 
print ('Decission Tree Classifier 1:')

# Fit
start = time.time()
dtree1.fit(X_train_pca,y_train)
finish = time.time()

# Prediction
y_pred = dtree1.predict(X_test_pca)

#prediction on training set
train_pred = dtree1.predict(X_train_pca)

# Checking Performance using different matrix
accuracy = accuracy_score(y_test, y_pred)
train_accuracy = accuracy_score(y_train, train_pred)
precision = precision_score(y_test,y_pred, average = 'macro')
recall = recall_score(y_test,y_pred, average = 'macro')
f1 = f1_score(y_test,y_pred, average = 'macro')

print('Training Accuracy: {:.3f}'.format(train_accuracy))
print('Test Accuracy: {:.3f}'.format(accuracy))
print('Precision: {:.3f}'.format(precision))
print('Recall: {:.3f}'.format(recall))
print('F1: {:.3f}'.format(f1))
print('Running time to fit the model: {:.3f} ms\n'.format((finish-start)*10**3))


# Decission tree classifier 2 
print ('Decission Tree Classifier 2:')

# Fit
start = time.time()
dtree2.fit(X_train_pca,y_train)
finish = time.time()

# Prediction
y_pred = dtree2.predict(X_test_pca)

#prediction on training set
train_pred = dtree2.predict(X_train_pca)

# Checking Performance using different matrix
accuracy = accuracy_score(y_test, y_pred)
train_accuracy = accuracy_score(y_train, train_pred)
precision = precision_score(y_test,y_pred, average = 'macro')
recall = recall_score(y_test,y_pred, average = 'macro')
f1 = f1_score(y_test,y_pred, average = 'macro')

print('Training Accuracy: {:.3f}'.format(train_accuracy))
print('Test Accuracy: {:.3f}'.format(accuracy))
print('Precision: {:.3f}'.format(precision))
print('Recall: {:.3f}'.format(recall))
print('F1: {:.3f}'.format(f1))
print('Running time to fit the model: {:.3f} ms\n'.format((finish-start)*10**3))


# Decission tree classifier 3 
print ('Decission Tree Classifier 3:')

# Fit
start = time.time()
dtree3.fit(X_train_pca,y_train)
finish = time.time()

# Prediction
y_pred = dtree3.predict(X_test_pca)

#prediction on training set
train_pred = dtree3.predict(X_train_pca)

# Checking Performance using different matrix
accuracy = accuracy_score(y_test, y_pred)
train_accuracy = accuracy_score(y_train, train_pred)
precision = precision_score(y_test,y_pred, average = 'macro')
recall = recall_score(y_test,y_pred, average = 'macro')
f1 = f1_score(y_test,y_pred, average = 'macro')

print('Training Accuracy: {:.3f}'.format(train_accuracy))
print('Test Accuracy: {:.3f}'.format(accuracy))
print('Precision: {:.3f}'.format(precision))
print('Recall: {:.3f}'.format(recall))
print('F1: {:.3f}'.format(f1))
print('Running time to fit the model: {:.3f} ms\n'.format((finish-start)*10**3))
```

## MNIST Dataset : Import and Transform

```python
# Load the MNIST dataset
X, y = fetch_openml("mnist_784", version=1, return_X_y=True, as_frame=False)

# Split the dataset to create a subset (3% of the data (train data) which is 2100 instance)
X_subset, _, y_subset, _ = train_test_split(X, y, test_size=0.97, random_state=1, stratify=y)

# Now we have a subset of the dataset (X_subset and y_subset)

# Split the subset into training and testing sets for the classifier
X_train, X_test, y_train, y_test = train_test_split(X_subset, y_subset, test_size=0.3, random_state=1, stratify = y_subset)

# Standardization

sc = StandardScaler()
X_train_std = sc.fit_transform(X_train)
X_test_std = sc.transform(X_test)

```
